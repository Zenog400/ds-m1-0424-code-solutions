```python
import sqlite3 as sq
import pandas as pd
```

```python
con = sq.connect("stocks.sqlite")
cur = con.cursor()
cur
```

```python
results = cur.execute("select name from sqlite_master where type='table'")
results.fetchall()
```

```python
results = cur.execute("pragma table_info(STOCK_DATA)")
db_meta = results.fetchall()
colNames = [x[1] for x in db_meta]
colNames
```

```python
ms = cur.execute("""select *
                 from STOCK_DATA
                 where Symbol = 'MSFT'""")
msFrame = pd.DataFrame(ms, columns=colNames)
msFrame
```

```python
apl = cur.execute("""select *
                 from STOCK_DATA
                 where Symbol = 'AAPL'""")
aplFrame = pd.DataFrame(apl, columns=colNames)
aplFrame
```
